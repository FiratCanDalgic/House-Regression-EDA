```python
original_image = np.random.randint(0, 256, size=(10, 10))
print("Orijinal Matris (Görüntü):\n", original_image)

def karart(image_array):
    # Tüm elemanlara aynı anda 2'ye bölme işlemi uygulanır
    return image_array / 2

    # Merkezi 4x4 alanı seç ve 255 ile çarp (parlak yap)
vurgulanmis_image = original_image.copy() # Orijinali korumak için kopya alın
vurgulanmis_image[3:7, 3:7] = 255 

print("\nVurgulanmış Merkez Matris:\n", vurgulanmis_image)

dondurulmus_image = original_image.T
print("\nDöndürülmüş (Transpoze) Matris:\n", dondurulmus_image)
```
